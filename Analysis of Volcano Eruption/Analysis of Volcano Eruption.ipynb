## query 5: Compute the mean years between eruptions per volcano during its known active span. 

<font size='3'>define a function to calculate this query</font>
<font size='3'>:param year - DataFrame</font>
<font size='3'>:param records - Series</font>

```python
def calActiveSpan(year,records):
    #add column'size' to dataframe year
    year['size']=records
    #add column'active span' to dataframe year
    year['active span']=np.NaN
    
    #when volcano only has one row, define its 'active span' as NaN
    singlerow=year['size']==1
    year.loc[singlerow,'active span']=np.NaN
    
    #when volcano has more than one row and its eruption is within a single year
    mutirow=year['size']>1
    oneyearspan=year['latest year']==year['ealiest year']
    year.loc[mutirow&oneyearspan,'active span']=1/year['size']
    
    #when volcano has more than one row and its eruption have yearspan
    mutiyearspan=year['latest year']>year['ealiest year']
    year.loc[mutirow&mutiyearspan,'active span']=(year['latest year']-year['ealiest year'])/year['size']
    
    return year
```

```python
yeardf=calActiveSpan(yeardf,records)
```
