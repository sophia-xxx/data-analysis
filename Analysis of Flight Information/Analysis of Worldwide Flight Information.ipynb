## Query 2

### a. Distinct Airports reach non-stop from two airports  (Airport’s Name, IATA Code, City, and Country.)

<font size=4>Load routes data and define its headers</font>

```python
name=['Airline','Airline ID','Source airport','Source airport ID',
      'Destination airport','Destination airport ID','Codeshare','Stops','Equipment']
```

```python
routes=pd.read_csv('routes.txt',sep=',',names=name)
routes.head(10)
```

<font size=4>Get destination airports first</font>

```python
source1=routes['Source airport']=='BMI'
source2=routes['Source airport']=='CMI'
stop=routes['Stops']==0
destination=routes[(source1|source2)&stop]
destination
```

<font size=4>Merge airports information table and the destination airports.<p>
Column 'Airport ID' must be an 'object' datatype, otherwise merging process will have mistakes.<p></font>

```python
mergetable=pd.merge(airports,destination,left_on='Airport ID',right_on='Destination airport ID',how='inner')
mergetable.head(5)
```

<font size=4>Get the columns we need.<p>

```python
destination_info=mergetable[['Airport ID','Name','IATA','City','Country']].drop_duplicates(subset='IATA')
```

```python
destination_info.set_index('Airport ID')
```

### b. The number of distinct Airports reach non-stop or one-stop from two airports  (Airport’s Name, IATA Code, City, and Country.)

<font size=4>Get destination airports first</font>

```python
source1=routes['Source airport']=='BMI'
source2=routes['Source airport']=='CMI'
stop=routes['Stops']==1
destination2=routes[stop&(source1|source2)]
destination2
```

<font size=4>Merge airports information table and the one-stop destination airports.</font>

```python
mergetable2=pd.merge(airports,destination2,left_on='Airport ID',right_on='Destination airport ID',how='inner')
mergetable2
```

```python
destination_info2=mergetable2[['Airport ID','Name','IATA','City','Country']].drop_duplicates(subset='IATA')
destination_info2.set_index('Airport ID')
```

```python
distinct_airport=pd.concat([destination_info,destination_info2],join='inner').set_index('Airport ID')
distinct_airport
```

## Query 5. Produce DataFrame: 
index-country(ascend)<p>
column:<p>
1) the number of distinct Airports in that country; <p>
2) the number of domestic routes listed [source and destination within the country]; <p>
3) the number of OUTbound international routes;(source country) <p>
4) the number of IN-bound international routes.(destination country) <p>
Hint: this is trickier than it seems at first even though we have grouping by Country in common. You may break it into several simpler queries, then join those correctly into your final result.<p>

<font size=4>Get each countries' distinct Airports number</font>

```python
country_airports=airports.groupby('Country').size()
country_airports=pd.DataFrame(country_airports)
country_airports.columns=['Airports number']
country_airports
```

<font size=4>Extract the columns we need from airports table and routes table</font>

```python
newroutes=routes[['Source airport ID','Destination airport ID']]
newroutes
```

```python
newairports=airports[['Airport ID','Country']]
newairports
```

<font size=4>Merge two tables: newairports table and newroutes table</font>

```python
first_merge=pd.merge(newroutes,newairports,left_on='Source airport ID',right_on='Airport ID',how='inner')
second_merge=pd.merge(first_merge,newairports,left_on='Destination airport ID',right_on='Airport ID',how='inner')
second_merge
```

<font size=4>Get domestic routes and group by 'Country'</font>

```python
domestic_routes=second_merge[second_merge['Country_x']==second_merge['Country_y']]
domestic_routes=pd.DataFrame(domestic_routes.groupby('Country_x').size())
domestic_routes.index.name='Country'
domestic_routes.columns=['Domestic routes number']
domestic_routes
```

<font size=4>Get the OUTbound and INbound routes(apart from the Domestic routes)</font>

```python
io_routes=second_merge[second_merge['Country_x']!=second_merge['Country_y']]
io_routes
```

<font size=4>Group the OUTbound routes by 'Source Airport Country' </font>

```python
outbound_routes=pd.DataFrame(io_routes.groupby('Country_x').size())
outbound_routes.index.name='Country'
outbound_routes.columns=['Outbound international routes']
outbound_routes
```

<font size=4>Group the INbound routes by 'Destination Airport Country'</font>

```python
inbound_routes=pd.DataFrame(io_routes.groupby('Country_y').size())
inbound_routes.index.name='Country'
inbound_routes.columns=['Inbound international routes']
inbound_routes
```

<font size=4>Merge the calculated airports, domestic, outbound and inbound routes by column 'Country'</font>

```python
firstmerge=pd.merge(country_airports,domestic_routes,left_index=True,right_index=True,how='outer')
firstmerge.head(10)
```

```python
secondmerge=pd.merge(firstmerge,outbound_routes,left_index=True,right_index=True,how='outer')
secondmerge.head(10)
```

```python
finalmerge=pd.merge(secondmerge,inbound_routes,left_index=True,right_index=True,how='outer')
finalmerge.head(10)
```
